```python
%%writefile module.py
def prime(x):
    '''Return True if number is prime otherwise False'''
    if x<2:
        print('Invalid Number')
    elif x==2:
        return True
    else:
        for i in range(2,x//2+1):
            if x%i==0:
                return False
        else:
            return True
def fact(x):
    '''Return the factorial of a number'''
    sum1=1
    for i in range(x,1,-1):
        sum1*=i
    return sum1
def add(x,y):
    '''return the addition of two given number'''
    return x+y
def sub(x,y):
    '''return the substraction of two given number'''
    return x-y
def mul(x,y):
    '''return the multiplication of two given number'''
    return x*y
```
